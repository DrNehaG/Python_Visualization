```python
# Importing the requisite libraries
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# To get the list of datasets available for practice in seaborn library
sns.get_dataset_names()
```

```python
# Loading the tips dataset
tips_df = sns.load_dataset('tips')
tips_df
```

```python
# To get the scatterplot
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips_df)
```

```python
# hue provides different colors to different categories for the chosen categorical variable
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips_df, hue = 'time')
```

```python
# style provides different shapes to the different categories for the chosen categorical variable
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips_df, hue = 'day', style='smoker')
```

```python
# Visualizing the data of Thursday only
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips_df[tips_df.day == 'Thur'], hue = 'smoker')
```

```python
# size provides different sizes of the chosen shape as per the values of the numerical variable
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips_df[tips_df.day == 'Thur'], hue = 'smoker', size = 'tip')
```

```python
# Creating a dataframe and getting the line chart
df = pd.DataFrame(np.random.randn(300,2) , columns=['a' , 'b'])
sns.relplot(x = 'a' , y = 'b' , kind = 'line' , data = df, sort = True )
```

**Remark:** By default relplot displays scatterplot.

```python
# Another way of getting the scatter plot
sns.scatterplot(x = 'tip', y = 'total_bill', data = tips_df)
```

```python
# Visualizing one numeric variable along with one categorical variable
sns.catplot(y = 'tip', x = 'day', data=tips_df)
```

```python
# Creating a boxplot for various categories of categorical variable along with one numerical variable
sns.catplot(y = 'tip', x = 'day', data=tips_df, kind='box')
```

```python
# To get the histogram of the numerical variable
sns.displot(x = 'tip', data = tips_df)
```

```python
# To get the joint plot corresponding to 2 numeric variables.
# It's showing the individual graph of both the variables as well as relationship between both these variables.
sns.jointplot(x = 'total_bill', y = 'tip', data = tips_df)
```

```python
# To see hexagon as the shape instead of circles in scatter plot that we get through joint plot.
sns.jointplot(x = 'total_bill', y = 'tip', data = tips_df, kind = 'hex')
```

```python
# To get the scatter plot for the combination of every 2 numeric variables in the dataset.
sns.pairplot(sns.load_dataset('iris'))
```

```python
# To get the line of best fit
# lmplot: linearmodelplot
sns.lmplot(x = 'total_bill', y = 'tip', data = tips_df)
```

```python
# Another way of getting the line of best fit
sns.regplot(x = 'total_bill', y = 'tip', data = tips_df)
```

```python
# To get the different line of best fit for each category of the categorical variable
sns.lmplot(x = 'total_bill', y = 'tip', data = tips_df, col='day')
```
